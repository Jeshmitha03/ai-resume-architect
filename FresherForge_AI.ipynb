```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# --- CONFIGURATION ---
# Replace with your actual key from Google AI Studio
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

# 1. Initialize the AI Brain
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.7)

# 2. The Refiner Engine (The "Secret Sauce")
# This turns "I built a calculator" into "Engineered a high-precision arithmetic tool..."
refine_template = ChatPromptTemplate.from_template("""
You are a Senior Technical Recruiter.
The candidate says: {student_input}
Rewrite this for a resume using the 'Action + Task + Result' framework.
Keep it to one high-impact bullet point.
""")
refine_chain = refine_template | llm | StrOutputParser()

# 3. The Interview Instructions
system_instructions = """
You are an AI Resume Coach. Collect Name, Education, and 1 Project.
After getting the project, stop and wait for the Refiner to work.
"""
```

```python
# --- THE LIVE INTERVIEW LOOP ---

print("=== AI RESUME PLATFORM v1.0 ALPHA ===")
name = input("Bot: Welcome! What is your full name? ")
edu = input(f"Bot: Nice to meet you, {name}. What is your Degree and College? ")
project_raw = input("Bot: Tell me about one technical project you worked on recently: ")

print("\n[AI is architecting your professional profile...]")

# Invoke the Refiner Chain
refined_bullet = refine_chain.invoke({"student_input": project_raw})

# Final Presentation Output
final_resume = f"""
-----------------------------------------
{name.upper()}
Education: {edu}
-----------------------------------------
TECHNICAL EXPERIENCE:
* {refined_bullet}
-----------------------------------------
"""

print("\n--- GENERATED RESUME PREVIEW ---")
print(final_resume)
```
